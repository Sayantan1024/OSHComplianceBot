```python
import json

with open("../data/processed/osh_sections.json", "r", encoding="utf-8") as f:
    data = json.load(f)

documents = [item["text"] for item in data]

print(f"✅ Loaded {len(documents)} OSH sections.")
print(documents[0][:500])  # preview
```

```python
from sentence_transformers import SentenceTransformer
embedder = SentenceTransformer('l3cube-pune/indic-sentence-similarity-sbert')
embeddings = embedder.encode(documents, show_progress_bar=True)
```

```python
import numpy as np
import faiss

# Convert to float32 (required by FAISS)
embeddings = np.array(embeddings).astype("float32")

# Create FAISS index
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)
```

```python
import pandas as pd

# Create a DataFrame from your original JSON data
df = pd.DataFrame(data)

# Add the new multilingual embeddings as a column
# We convert the numpy arrays back to standard Python lists so they save correctly in the CSV
df['embedding'] = embeddings.tolist()

# Define a NEW filename to prevent overwriting your old work
# Adjust the path if you are running this in Colab instead of locally
output_path = "../data/processed/osh_sections_multilingual_(1)_vectors.csv"

# Save the DataFrame to the new CSV
df.to_csv(output_path, index=False)

print(f"✅ Successfully saved multilingual vectors to: {output_path}")
```
